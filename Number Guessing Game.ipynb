```python
import random
num, attempts = random.randint(1, 100), 10
while attempts > 0:
    guess = int(input("Enter a number: "))
    if guess == num:
        print("Correct! You guessed the number.")
        break
    elif guess < num:
        print("Too low.")
    else:
        print("Too high.")
    attempts -= 1
if attempts == 0:
    print(f"Out of attempts! The number was %d." % num)
```
