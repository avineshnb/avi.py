```python
num = int(input("Enter a number: "))
for x in range(num + 1):
    if x % 2 == 0:
        parity = "even"
    else:
        parity = "odd"
    is_prime = True
    if x <= 1:
        is_prime = False
    else:
        for i in range(2, int(x ** 0.5) + 1):
            if x % i == 0:
                is_prime = False
    a, b = 0, 1
    is_fibonacci = (x == 0)
    while b <= x:
        is_fibonacci = is_fibonacci or (x == b)
        a, b = b, a + b
    properties = [parity]
    if is_prime:
        properties.append("prime")
    if is_fibonacci:
        properties.append("fibonacci")
    print(f"{x} - {', '.join(properties)}")
```
