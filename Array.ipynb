```python
nums = []
for x in range(10):
  num = int(input("Enter a number: "))
  nums.append(num)
for i in range(10):
  for k in range(i + 1, 10):
    if nums[i] > nums[k]:
      nums[i], nums[k] = nums[k], nums[i]
print("Sorted numbers:", nums)
```

```python
nums = []
for x in range(10):
  num = int(input("Enter a number: "))
  nums.append(num)
minimum = nums[0]
for x in range(len(nums)):
  if minimum > nums[x]:
    minimum = nums[x]
print("Min: %d" % minimum)
```
