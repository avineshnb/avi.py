```python
import random
nums = list(range(1, 11))
random.shuffle(nums)
print("Shuffled:")
for x in nums:
    print(x)
nums.sort()
print("\nSorted Accending:")
for x in nums:
    print(x)
```

```python
import random
nums = list(range(1, 11))
random.shuffle(nums)
print("Shuffled:")
for x in nums:
    print(x)
nums.sort(reverse=True)
print("\nSorted Descending:")
for x in nums:
    print(x)
```
