```python
num, parity, sum = int(input("Enter a number: ")), input("Enter even or odd: "), 0
if parity.casefold() == "even":
  for x in range(num + 1):
    if x % 2 == 0:
      sum += x
  print("The sum of all even numbers from 0 to %d is %d" % (num, sum))
elif parity.casefold() == "odd":
  for x in range(num + 1):
    if x % 2 != 0:
      sum += x
  print("The sum of all odd numbers from 0 to %d is %d" % (num, sum))
else:
  print("Not a valid parity")
```
